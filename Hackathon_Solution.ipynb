# Data Analysis for xyz vaccine

```python
columns_for_xyz = ['xyz_concern', 'xyz_knowledge', 'behavioral_avoidance', 'behavioral_face_mask', 'behavioral_wash_hands', 'behavioral_large_gatherings',
               'behavioral_outside_home', 'behavioral_touch_face', 'doctor_recc_xyz', 'chronic_med_condition', 'child_under_6_months', 
               'health_worker', 'opinion_xyz_vacc_effective', 'opinion_xyz_risk', 'opinion_xyz_sick_from_vacc', 
               'age_group', 'hhs_geo_region', 'census_msa', 'household_adults', 'household_children']
data_xyz = data[columns_for_xyz]
data_xyz.isna().sum()
```

```python
data_xyz.shape
```

```python
columns_to_NAN = ['behavioral_face_mask', 'behavioral_wash_hands', 'behavioral_large_gatherings', 'behavioral_outside_home', 'behavioral_touch_face', 
               'behavioral_avoidance', 'chronic_med_condition', 'child_under_6_months', 'health_worker', 'doctor_recc_xyz', 
                'household_adults', 'household_children', 'xyz_concern', 'xyz_knowledge']
data_xyz[columns_to_NAN] = data_xyz[columns_to_NAN].fillna(0)
```

```python
data_xyz.isna().sum()
```

```python
data_xyz.dtypes
```

```python
data_xyz = pd.get_dummies(data_xyz, drop_first = True)
data_xyz.isna().sum()
```

```python
data_xyz['opinion_xyz_vacc_effective'] = data_xyz['opinion_xyz_vacc_effective'].fillna(data_xyz['opinion_xyz_vacc_effective'].mean())
data_xyz['opinion_xyz_risk'] = data_xyz['opinion_xyz_risk'].fillna(data_xyz['opinion_xyz_risk'].mean())
data_xyz['opinion_xyz_sick_from_vacc'] = data_xyz['opinion_xyz_sick_from_vacc'].fillna(data_xyz['opinion_xyz_sick_from_vacc'].mean())
```

```python
data_xyz.isna().sum()
```

# Logistic Regression for xyz vaccine

```python
train_x_1, test_x_1, train_y_1, test_y_1 = train_test_split(data_xyz, y_label['xyz_vaccine'], test_size = 0.25, random_state = 27)
```

```python
logic = LogisticRegression()
logic.fit(train_x_1, train_y_1)
```

```python
logic.score(test_x_1,test_y_1)
```

```python
xyz_vaccine = logic.predict_proba(data_xyz)[:, 1] 
```
